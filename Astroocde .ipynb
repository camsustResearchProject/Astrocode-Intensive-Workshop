<h3> Calculation </h3>
In the following codeblock we will make a simple calculator that will take two numbers and calculate their sum, their difference, their multiplication, their division and the remainder left after division.

```python
num1 = float(input("Enter a number: "))
num2 = float(input("Enter another number: "))

add = num1 + num2
sub = num1 - num2
prod = num1 * num2
div = num1 / num2
remainder = num1 % num2      # % is called the modulus operator. It returns the reminder after dividing num1 by num2

print(f"{num1} + {num2} = {add}")
print(f"{num1} - {num2} = {sub}")
print(f"{num1} * {num2} = {prod}")
print(f"{num1} / {num2} = {div}")
print(f"{num1} % {num2} = {remainder}")
```

### For Loop

Repating the same task again and again is what a loop does. Loops will always run until a condition is met. The syntax for loop in python programming language is 

`for i in range(iter_no):`
<br>
`    code`

the for loop uses `for` keyword. Then comes the `i` that counts how many times the loop was running. It is sometimes called a counter. then comes the in keyword followed by `range` function. The range function has the syntex `range(start,stop,step)`. And at last the loop statement begins the block using a colon (`:`).
<img src='FlowChartLoop.png'>

```python
for i in range(10):
    print("Hello")
```
